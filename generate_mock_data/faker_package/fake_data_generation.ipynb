```python
from faker import Faker
import numpy as np
import pandas as pd
import json
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
fake = Faker(locale='in')
```

```python
def make_workers(num):
    status_list = ['full time', 'part time', 'per post']
    team_list = [fake.color_name() for x in range(5)]
    fake_workers = [
        {
            'worker id' : x+1000,
            'worker name' : fake.name(),
            'hire date' : fake.date_between(start_date='-2y', end_date='today'),
            'worker status' : np.random.choice(status_list, p=[0.5, 0.3, 0.2]),
            'team' : np.random.choice(team_list)
        }
        for x in range(num)
    ]
    
    return fake_workers
```

```python
workers_df = pd.DataFrame(make_workers(num=5000))
```
