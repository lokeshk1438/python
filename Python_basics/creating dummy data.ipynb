```python
import pandas as pd
from faker import Faker
import numpy as np
```

```python
fake = Faker()
```

```python
dir(fake)
```

```python
for _ in range(10):
    print(fake.credit_card_number())
```

```python
cards=[]
```

```python
for _ in range(100000):
    card=fake.credit_card_number()
    if len(card) == 16:
        cards.append(card)
```

```python
names=[]
```

```python
for _ in range(50021):
    names.append(fake.first_name_male())
```

```python
df_cards=pd.DataFrame(cards)

df_names=pd.DataFrame(names)
```

```python
df = pd.DataFrame()

df = pd.concat([df_cards, df_names], axis=1, ignore_index=True)

df.columns=['Card_Names', 'Names']
```

```python
df.to_csv('cards_names.csv', index=False)
```
