```python
for dtype in ['float', 'int', 'object']:
    selected_df = g1.select_dtypes(include=dtype)
    mem_usage_b = selected_df.memory_usage(deep=True).mean()
    mem_usage_mb = mem_usage_b / 1024 ** 2
    print("Average memory usage for {} columns: {:03.2f} MB".format(dtype,mem_usage_mb))
```

```python
# checking min and max values of a data-type

int_types = ['int8', 'int16', 'uint8']
for itype in int_types:
    print(np.iinfo(itype))
```

```python
def mem_usage(pandas_obj):
    if isinstance(pandas_obj, pd.DataFrame):
        mem_usage_b = pandas_obj.memory_usage(deep=True).mean()
    else:
        mem_usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = mem_usage_b / 1024 ** 2
    return "{:.2f} MB".format(usage_mb)
```

## conversion to int

```python
g1_int = g1.select_dtypes(include=['int'])
converted_int = g1_int.apply(pd.to_numeric, downcast='unsigned')
print(mem_usage(g1_int))
print(mem_usage(converted_int))

compare_ints = pd.concat([g1_int.dtypes, converted_int.dtypes], axis=1)
compare_ints.columns = ['before', 'after']
```
