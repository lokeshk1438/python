## Row Number within each group (partition)

```python
# The Pandas equivalent of row number within each partition with multiple sort by parameters:

# ROW_NUMBER() over (PARTITION BY ticker ORDER BY date DESC) as days_lookback
# # ---------
# ROW_NUMBER() over (PARTITION BY ticker, exchange ORDER BY date DESC, period) as example_2
```

```python
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
```

```python
df = pd.read_csv(url)

df = df.dropna()

df = df.rename(columns=str.lower).reset_index(drop=True)
```

```python
df.head(10)
```

```python
# finding the smallest person in each gender

(
    df.assign(
        rn = df.sort_values('age')
        .groupby('sex')
        .cumcount() + 1
    )
    .sort_values(['sex', 'rn'])
    .reset_index(drop=True)
)
```
